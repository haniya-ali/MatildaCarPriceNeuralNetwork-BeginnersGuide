## 1. An Introduction to Neural Networks

### Definition of Neural Networks

While **machine Learning** is a sub-field of AI that focuses on building algorithms that learn from experience, primarily the data they are exposed to, "an **artificial neural network** is a machine learning algorithm based on the concept of a human neuron". (https://www.ncbi.nlm.nih.gov/pmc/articles/PMC6428006/)


### Basic Structure of Neural Networks

To understand and appreciate neural networks, it's worthwhile to delve into the fundamental building block they emulate: the human neuron. Much like the neurons in our brains, artificial neural networks consist of interconnected nodes that process and transmit information.

A human neuron comprises three main parts: **the dendrites, the cell body (soma), and the axon**. 

**Dendrites** act as receivers, collecting signals from other neurons and transmitting them to the cell body. 

**The cell body** processes these signals and, if a certain threshold is reached, sends an output signal down the **axon** to communicate with other neurons.


Image: A biological and an artificial neuron (via https://www.quora.com/What-is-the-differences-between-artificial-neural-network-computer-science-and-biological-neural-network)

The structure of the human neuron serves as the inspiration for artificial neural networks. In these computational models, nodes (artificial neurons) are organized into layers, each layer playing a specific role in processing and transmitting information. 

The main components of a neural network in terms of the anatomy of a neuron are as follows:

**1. Dendrites (receivers):** In an artificial neural network, the **input layers** are similar to dendrites. They serve as receivers that 
collect information, just as dendrites gather signals from other neurons in a biological neuron.

**2. Cell Body (processor):** Since the processing happens in the cell body, in a neural network, this corresponds to the **hidden layers** where computations take place. The cell body interprets the information and decides whether to pass it on.

**3. Axon (transmitter):** In a human neuron, if the processed information meets a certain threshold, it moves down the axon. In the artificial network, this is similar to the **output layer**, where the final decision or result is either transmitted to the next layer or is the ultimate output.

Additionally, there are the following as well:

**4. Synaptic Connections (weights):** In both biological and artificial neurons, the connections matter, and the strength of connections between neurons is like the **weights** in our digital network. These weights determine the influence each input has on the final decision.

**5. Activation Function (decision Rule):** Finally, the decision is made through an **activation function**, representing the rule our digital neuron follows. If the result is above a certain threshold, it activates and passes the information.

### More on activation functions

Activation function serve the purpose of introducing a level of complexity and flexibility to neural networks. They add non-linearity to the network, enabiling it to learna dn adapt to intricate patterns in the data. Without activation functions, the networks would operate in a linear fashion. Since many real-world relationships are not linear or simple, activation functions introduce non-linearity by adding curves and bends to the decision-making process.

After understanding their importance, lets discuss some of the most popular types of Activation Functions:

**a. Binary Step Function:**

    Definition: The binary step function is the simplest activation function, outputting either 0 or 1 based on a threshold.
    Purpose: It can be useful in scenarios where a binary decision needs to be made.

**b. Linear Activation Function:**

    Definition: The linear activation function outputs values in a straight line, directly proportional to the input.
    Purpose: It is often used in simple tasks, but less common in hidden layers, as it lacks the non-linearity crucial for handling complex patterns.

**c. Non-linear Activation Functions (e.g., tanh, sigmoid, ReLU):**

    Definition: Activation functions that introduce non-linearity by adding curves and bends to the decision-making process.
    Purpose: Enable the network to learn and adapt to intricate patterns in data, crucial for handling the complexity of real-world datasets.

**d. Sigmoid Function:**

    Definition: The sigmoid function squashes input values between 0 and 1.
    Purpose: Commonly used in the output layer of binary classification tasks, where the goal is to make a decision (yes/no).
    
**e. Hyperbolic Tangent (tanh):**

    Definition: Similar to the sigmoid but squashes input values between -1 and 1.
    Purpose: Often used in the hidden layers of the network to handle a wider range of input data.
    
**f. Rectified Linear Unit (ReLU):**

    Definition: Outputs the input directly if it's positive; otherwise, it outputs zero.
    Purpose: Popular in hidden layers due to its simplicity and effectiveness. It helps the network learn faster and can handle complex patterns.
    
**g. Softmax Function:**

    Definition: Converts a vector of numbers into probabilities, making it suitable for multi-class classification.
    Purpose: Used in the output layer of a neural network when dealing with multiple classes.




### Feedforward and backpropagation

"A feedforward neural network is one of the simplest types of artificial neural networks devised. In this network, the information moves in only one directionâ€”forwardâ€”from the input nodes, through the hidden nodes (if any), and to the output nodes." (https://deepai.org/machine-learning-glossary-and-terms/feed-forward-neural-network

The components of a feedforward neural network are **the input layer, hidden layers, and the output layer**. 

Feedforward neural networks work in two phases: **the feedforward phase and the backpropagation phase**.

**Feedforward Phase:**
Imagine a neural network as a fantasy adventure. The story begins with the protagonists (input data) going on a journey through the book's pages (the layers of the network) to discover their destiny. Along the way, each page (hidden layer) presents them with novel challenges, shaping their journey. Some pages contain significant twists and turns (activation functions) that hold the key to the protagonists' fate. This captivating journey continues until the characters reach the concluding page (output layer), where their destiny is revealed.

**Backpropagation Phase:**
Now, consider the backpropagation phase as the characters reflect on their adventure to enhance its enchantment. Following the grand revelation (prediction), the characters compare their tale with the actual outcome (actual output). Identifying disparities in their story (error), the characters share this insight backward through the pages. Each storyteller then adjusts their part of the tale, refining the narrative. This iterative editing process, guided by a special recipe (gradient descent algorithm), unfolds repeatedly, adding layers of excitement with each round.

So, in simple terms, the feedforward phase is like telling a fantasy story, and the backpropagation phase is the characters reviewing and editing to make the story even more enchanting and accurate. 







### Other types of neural networks

Apart from feedforward neural networks, some other types of neurak networks are as follows:

1. Recurrent Neural Networks (RNN): RNNs have connections that form loops, allowing them to process sequential data by maintaining a memory of previous inputs. They are ideal for tasks involving sequences, such as natural language processing, time series analysis, and speech recognition.
   
2. Convolutional Neural Networks (CNN):CNNs are designed to efficiently process grid-like data, such as images. They use convolutional layers to automatically and adaptively learn spatial hierarchies of features. They are widely used in image and video recognition, object detection, and classification tasks.
   
3. Long Short-Term Memory Networks (LSTM): An extension of RNNs, LSTMs are specialized in learning long-term dependencies in data. They are valuable for tasks where understanding context over longer sequences is crucial, such as language translation and speech recognition.


### Training and Learning in Neural Networks:



Training a neural network is an iterative learning process where initially random weights and biases are adjusted through multiple cycles of prediction, error calculation, and backpropagation. In each iteration, the network makes predictions, assesses the error using a loss function, and fine-tunes its parameters through backpropagation to minimize errors. This iterative learning, guided by optimization algorithms like gradient descent, continues until the model becomes well-trained.

### Importance and Applications

Neural networks can model very complex relationships which makes them ideal for the following tasks:
1. Image recognition
2. Natural language processing
3. Predictive modeling

### Challenges

Although neural networks are quite powerful, they come with a set of challenges such as the following:
1. Parameter tuning: "One of the main challenges is the choice of the number of hidden layers and the number of neurons in each layer, which can significantly affect the performance of the network." (https://deepai.org/machine-learning-glossary-and-terms/feed-forward-neural-network)
   
2. Large Datasets: Neural networks often require large datasets for effective training, which can be a limitation in scenarios where obtaining extensive labeled data is challenging or expensive.
   
3. Potential Overfitting: There's a risk of overfitting, where the model becomes too specialized in the training data and performs poorly on new, unseen data. Balancing complexity and generalization is crucial to mitigate overfitting.

4. Interpretability: The complexity of neural networks can hinder interpretability, making it challenging to understand how the model reaches specific decisions

## 5. Training the model

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size = 0.25)
```

```python
import tensorflow
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import MinMaxScaler

model = Sequential()
#Adding a hidden layer with 25 neurons and ReLU activation, specifying the input dimension as 5.
model.add(Dense(25, input_dim=5, activation='relu'))

#Adding another hidden layer with 25 neurons and ReLU activation.
model.add(Dense(25, activation='relu'))

#Adding an output layer with 1 neuron and linear activation.
model.add(Dense(1, activation='linear'))

model.summary()
```

```python
# Compiling the model with Adam optimizer and mean squared error loss function
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# epochs=20: Number of times the entire training dataset is passed forward and backward through the neural network
# batch_size=25: Number of samples per gradient update, helps control the optimization process
# verbose=1: Display training progress during each epoch
# validation_split=0.2: Fraction of the training data to be used as validation data, in this case, 20% of the data
epochs_hist = model.fit(X_train, y_train, epochs=20, batch_size=25,  verbose=1, validation_split=0.2)
```

## 6. Model Evaluation

```python
print(epochs_hist.history.keys())
```

The code below uses Matplotlib to create a line plot of the training and validation loss over epochs. It helps visualize how well the model is learning from the training data and how well it generalizes to unseen validation data. The legend distinguishes between the training and validation loss on the plot.

```python
# Plotting the training and validation loss over epochs
# epochs_hist.history['loss']: Training loss values stored during training
# epochs_hist.history['val_loss']: Validation loss values stored during training
plt.plot(epochs_hist.history['loss'], label='Training Loss')
plt.plot(epochs_hist.history['val_loss'], label='Validation Loss')

# Setting plot title and axis labels
plt.title('Model Loss Progression During Training/Validation')
plt.ylabel('Training and Validation Losses')
plt.xlabel('Epoch Number')

# Displaying a legend to distinguish between training and validation losses
plt.legend(['Training Loss', 'Validation Loss'])
```
